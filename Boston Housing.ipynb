## Assumptions of Linear Regression


### There are 5 Basic assumptions of Linear Regression Algorithms

#### 1. Linear Relationship between the features.
   Linear relationship between the features and target.
#### 2. Little or no Multicollinearity between the features.
   Multicollinearity is a state of very high inter-correlation or inter-associations among the independent variables. It is therefore a type of distribance in the data present weakns the statistical power of the regression model.Pair plots and heatmaps(correlation matrix)can be used for identifying highlybcorrelated features.
#### 3. No or Low Heteroscadasticity 
   Error term should not explain other error term behavior when plotted scatter plot of errors they should random spread instead of showing any pattern.
#### 4. Normal Distributiion

#### 5. Little or No autocorrelation in the residuals
